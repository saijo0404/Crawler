```python
import asyncio
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.common.by import By
import time
import requests
import os
import base64
from io import BytesIO
from PIL import Image

path = 'save_path'

def download_image(image_url, path, name):
    if not os.path.exists(path):
        os.makedirs(path)
        
    name_format = "01"
    fullname = str(name).zfill(len(name_format))+".jpg"
    if not os.path.exists(path+fullname):
        try:
            # Extract the Base64-encoded data
            image_data = image_url.split(',')[1]
            # Decode the Base64 data
            binary_data = base64.b64decode(image_data)
            # Create a BytesIO object from the binary data
            image_stream = BytesIO(binary_data)
            # Open the image using PIL (Python Imaging Library)
            image = Image.open(image_stream)
            # Save the image to a file
            image.save(path+fullname)
            # Display the image
            # image.show()
        except Exception as e:
            print(e)
        
# Edge WebDriver 的路徑，根據你的系統和 WebDriver 的位置進行修改
edge_driver_path = './msedgedriver.exe'
# 創建 Edge 瀏覽器實例
options = webdriver.EdgeOptions()
options.add_argument("--mute-audio")  # 設定瀏覽器静音
# 啟動 Edge 瀏覽器
driver = webdriver.Edge(service=EdgeService(executable_path=edge_driver_path), options=options)

# https://kep1er.jp/photo 找所有貼文連結
links = []
for i in range(1,4):
    url = 'https://kep1er.jp/photo?page='+str(i)
    driver.get(url)
    time.sleep(5)
    href = driver.find_elements(By.XPATH, "//a[contains(@href, '/photo/')]")
    for h in href:
        links.append(h.get_attribute('href'))
# print(links)
print(len(links))


# 登入
url = 'https://kep1er.jp/login'
driver.get(url)
time.sleep(5)
inputs = driver.find_elements(By.TAG_NAME, 'input')
account_info = ['account', 'password']
for index, send in enumerate(inputs):
    send.send_keys(account_info[index])
buttons = driver.find_elements(By.TAG_NAME, 'button')
buttons[1].click()

titles = []
for url in links:
    driver.get(url)
    time.sleep(3)
    title = driver.find_element(By.XPATH, '//*[@id="styled-root"]/main/section/h1').text
    title = title.replace("<", "").replace(">", "").replace("\\", "").replace("/", "")
    temp = title
    if title in titles:
        title = title + '_'+str(titles.count(title))
    titles.append(temp)
    images = driver.find_elements(By.XPATH, "//img[contains(@src, 'blob:')]")
    print(titles[-1])
    for count, image in enumerate(images):
        try:
            blob_url = image.get_attribute('src')
            # 使用 JavaScript 執行程式，將 Blob URL 轉換為實際圖片 URL
            image_url = driver.execute_async_script('''
                const callback = arguments[arguments.length - 1];
                var xhr = new XMLHttpRequest();
                xhr.open("GET", arguments[0], true);
                xhr.responseType = "blob";
                xhr.onload = function() {
                    var blob = xhr.response;
                    var reader = new FileReader();
                    reader.onloadend = function() {
                        callback(reader.result);
                    };
                    reader.readAsDataURL(blob);
                };
                xhr.send();
            ''', blob_url)
        finally:
            download_image(image_url, path+title+'/', count+1)
    
driver.quit()
print('end')
```
