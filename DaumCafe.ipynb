# 自動取得所有貼文影片

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.common.by import By
import urllib.request
import time
import os
import json
import requests
import shutil

path = 'save_path'

def download_image(url, path, name):
    if not os.path.exists(path):
        os.makedirs(path)
    test = "01"
    # Convert back to a string and use zfill to pad with zeros
    fullname = str(name).zfill(len(test))+".jfif"
    if not os.path.exists(path+fullname):
        try:
            urllib.request.urlretrieve(url,path+fullname)
        except Exception as e:
            print('not found')
            return
    
def download_video(url, path):
    if not os.path.exists(path):
        os.makedirs(path)
    r = requests.get(url, stream=True)
    with open(path+'01.mp4', 'wb') as f:
        shutil.copyfileobj(r.raw,f)
        f.close()
    
# Edge WebDriver 的路徑，根據你的系統和 WebDriver 的位置進行修改
edge_driver_path = './msedgedriver.exe'
# 創建 Edge 瀏覽器實例
options = webdriver.EdgeOptions()
options.add_argument("--mute-audio")  # 設定瀏覽器静音
# 啟動 Edge 瀏覽器
driver = webdriver.Edge(service=EdgeService(executable_path=edge_driver_path), options=options)

# 設定所需要 cookie
payload = {
    'webid': '',
    'webid_ts': '',
    'HM_CU': '',
    'ENAI': '',
    'TS': '',
    'HTS': '',
    'PROF': '',
    'ALID': '',
    'LSID': '',
    'alimcnt1ZBuS': '',
    'VISIT_FBLYN': '',
    '_T_ANO': '',
    'webid_sync': ''
}

# 訪問 Daum 카페: 추천 頁面並添加 cookie
url = 'https://cafe.daum.net/***'
driver.get(url)
for name, value in payload.items():
    driver.add_cookie({'name': name, 'value': str(value), 'domain': '.daum.net'})

# 添加完 cookie 訪問頁面
url = ''
driver.get(url)
time.sleep(2)

# 切換至 iframe 當中
iframe_id = 'down'
iframe = driver.find_element(By.ID, iframe_id)
driver.switch_to.frame(iframe)
    
links = []
curr_title_num = 0
title_class = 'title_wrapper' # get title
next_button = 'link_num'
count = 2
while True: 
    titles = driver.find_elements(By.CLASS_NAME, title_class)
    if len(links)==curr_title_num and curr_title_num!=0:
        break
    curr_title_num = len(links)
    for title in titles:
        href_class = 'txt_item'
        if title.find_element(By.CLASS_NAME, href_class).get_attribute('href') not in links:
            links.append(title.find_element(By.CLASS_NAME, href_class).get_attribute('href'))
    button = driver.find_elements(By.CLASS_NAME, next_button)
    for b in button:
        if b.text==str(count):
            b.click()
            count = count +1
            break
    time.sleep(2)
print(len(links))

for link in links:
    driver.get(link)
    time.sleep(2)
    iframe_id = 'down'
    iframe = driver.find_element(By.ID, iframe_id)
    driver.switch_to.frame(iframe)
    image_class = 'txc-image'
    video_class = 'figure-video'
    title_class = 'tit_info'
    title = driver.find_element(By.CLASS_NAME, title_class).text
    title = title.replace('/', '').replace('|', '').replace('\\', '').replace('?', '')
    print(title)
    
    try:
        video = driver.find_element(By.CLASS_NAME, video_class)
    except Exception as e:
        try:
            images = driver.find_elements(By.CLASS_NAME, image_class)
        except Exception as e:
            print('not found')
            continue
        if len(images)==0:
            continue
        # find image
        count = 1
        for image in images:
            download_image(image.get_attribute('src'), path+title+'/', count)
            count = count + 1
        continue
    # find video
    iframe = driver.find_elements(By.TAG_NAME, 'iframe')
    print(iframe[1].get_attribute('src'))
    download video
    download_video(iframe[1].get_attribute('src'), path+title+'/')
driver.quit()
print('end')
```
